```python
import math
import re

def integrate(xmax,nx,e1,psi):
	ee = e1
	h = xmax/nx
	h2 = h**2
	h12 = h2/12.0

	setinitcond(xmax,h,psi(-nx),psi(-nx+1))
	for i in range(-nx+2,nx):
		x = i*h
		numerovstep(x)
		psi[i] = p1

	normalize(psi,nx,h)

	return integrate(xmax,nx,e1,psi)

def numerovstep(x):
	q2 = h2*f1*p1+2.0*q1-q0
	q0 = q1
	q1 = q2
	f1 = 2.0*(potential(x)-ee)
	p1 = q1/(1.0-h12*f1)

	return numerovstep(x)

def potential(x):
	# square well with infinite walls
	potential = 0.0

	# square well with soft walls V(|x|>1) = 10
	#if abs(x) < 1.0:
	#	potential = 0.0
	#else:
	#	potential = 10.0

	# square well with hard walls internal and box between -0.5 and 0.5
	#if x > -1.50 and x < 1.50:
	#	potential = 1.0
	#else:
	#	potential = 0.0

	return potential(x)

def setinitcond(xmax,h,psi0,psi1):
	psi0 = 0.0
	psi1 = 0.00010

	p1 = psi0
	f1 = 2.0*(potential(-xmax)-ee)
	q0 = psi0*(1.0-h12*f1)
	f1 = 2.0*(potential(-xmax+h)-ee)
	q1 = psi1*(1.0-h12*f1)

	return setinitcond(xmax,h,psi0,psi1)

def normalize(psi,nx,h):
	norm = psi(-nx)**2+psi(nx)**2
	for j in range(-nx+1,nx-3,2):
		norm = norm+4.0*psi(j)**2+2.0*psi(j+1)**2
	norm = norm+4.0*psi(nx-1)**2
	norm = 1.0/math.sqrt(norm*h/3.0)
	for k in range(-nx,nx):
		psi[k] = psi[k]*norm

	return normalize(psi,nx,h)

def writemessage(n,e,b):
	print(n,'	E =' ,'{:,.7f}'.format(e),'	Boundary deviation  ','{:,.7f}'.format(b))

	return writemessage(n,e,b)


eps = 10**-6

xmax = float(input('Integrating between -xmax and xmax; give xmax\n'))
nx = float(input('Number of points on either side of 0\n'))
e1 = float(input('Searching for E>E0; give E0\n'))
de = float(input('Delta-E in initial course search\n'))

ni = 0
integrate(xmax,nx,e1,de)
b1 = psi(nx)
writemessage(ni,e1,b1)

while b1*b2 > 0.0:
	ni = ni+1
	e2 = e1+de
	integrate(xmax,nx,e2,psi)
	b2 = psi(nx)
	writemessage(ni,e2,b2)
	e1 = e2
	b1 = b2

ni = 0
while abs(b0) >= eps:
	ni = ni+1
	e0 = (e1+e2)/2.0
	integrate(xmax,nx,e0,psi)
	b0 = psi(nx)
	writemessage(ni,e0,b0)
	if b0*b1 <= 0.0:
		e2 = e0
		b2 = b0
	else:
		e1 = e0
		b1 = b0

with open('psi_input.dat', 'w+') as file:
    data = file.read()
    match = re.findall('Li\s+\w+\s+\d+\.\d+\s+\d\.\d+\s+\d\.\d+\s+\d\s+\d\s+\d\s+\d', data)
    for mm in match:
        data = data.replace(mm, f'{mm[:-1]}1')
    file.write(data)
```
